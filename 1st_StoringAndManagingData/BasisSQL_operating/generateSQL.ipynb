```python
def generate_sql_sentence(table_name, len=200, insert_statement=lambda i: i, columns='') :
  sql = 'INSERT INTO "{}" {} \r\nVALUES '.format(table_name, columns)
  for i in range(1, len+1):
    sql += insert_statement(i)
  return sql
generate_sql_sentence('ExperimentalUnit', 200, lambda i:"({}, 'sample{}'){}".format(i, i, ", "if i<200 else ';'))
```
