## turning Celsius temperature int fahrenheit temperature

### pseudocode
* input the temperature as degrees Celsius
* calculate Fahrenheit as (9/5)Celsius + 32
* Out Fahrenheit

### code

```python
def convert_temperature():
  celsius = eval(input('What is the Celsius temperature? '))
  fahrenheit = 9/5 * celsius + 32
  print("The temperature is", fahrenheit, "degrees Fahrenheit.")
  # return fahrenheit
convert_temperature()
```

## algorithms exercise week 1
### specification
this snippet is try to exercise visualising the cost function, using scatter, landscape surface, contour to plot.
### code

```python
# simple linear regression
# scatter
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame([
                    [2, 3], 
                    [4, 2], 
                    [5, 4], 
                    [8, 7]
])
def univariate_linear_regression(theta0, theta1, X):
  return theta0 + theta1 * X

def J(m, y_prediction, Y):
  return (1 / (2 * m)) * sum((y_prediction - Y)**2)

def print_result(thetas, J):
  str = ''
  for i, theta in enumerate(thetas):
    str += 'theta{} is {}, '.format(i, theta)
  str += 'the Squared Error is {}'.format(J)
  print(str)
  

X = data.iloc[:, 0].values
Y = data.iloc[:, 1].values
m = float(len(X))
y_prediction1 = univariate_linear_regression(1, 2, X)
J1 = J(m, y_prediction1, Y)
print_result([1, 2], J1)

y_prediction2 = univariate_linear_regression(0, 1, X)
J2 = J(m, y_prediction2, Y)
print_result([0, 1], J2)

fig = plt.figure()
plt.scatter(X, Y, color='m', marker='o', s=30)
plt.plot(X, y_prediction1, color='red')
plt.plot(X, y_prediction2, color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('simple linear regression')
fig.savefig('simple linear regression exercise.png')
plt.show()
```

```python
# landscape surface plot, use different theta0 and theta1
import numpy as np
from matplotlib import  cm

theta0 = np.arange(.4, 2, .01)
theta1 = np.arange(.6, 1, .01)
X, Y = np.meshgrid(theta0, theta1)
data = pd.DataFrame([
                    [2, 3], 
                    [4, 2], 
                    [5, 4], 
                    [8, 7]
])
# Z: calculate cost function in the distribution of theta0 and theta1
Z = 1/8 * sum([ (X + row[0] * Y - row[1])**2 for row in data.values])

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax.set_xlabel('$\\theta_0$')
ax.set_ylabel('$\\theta_1$')
ax.set_zlabel('Cost Function $J(\\theta_0, \\theta_1)$')
ax.set_zlim(0.5, 5)
plt.title('Simple Linear Function $\\theta_0, \\theta_1, J(\\theta_0, \\theta_1)$')
fig.savefig('plot_surface.png')
plt.show()
```

```python
# contour plot
import numpy as np
import matplotlib.pyplot as plt

delta = 0.002
theta0 = np.arange(.4, 2, delta)
theta1 = np.arange(.6, 1, delta)
X, Y = np.meshgrid(theta0, theta1)
data = pd.DataFrame([
                    [2, 3], 
                    [4, 2], 
                    [5, 4], 
                    [8, 7]
])
# Z cost function theta0theta1
Z = 1/8 * sum([ (X + row[0] * Y - row[1])**2 for row in data.values])


fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z, 40, cmap='RdGy')
ax.clabel(CS, inline=True, fontsize=10)
ax.set_xlabel('$\\theta_0$')
ax.set_ylabel('$\\theta_1$')
plt.savefig('contour.png')
plt.show()
```
